```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from sklearn.metrics import silhouette_score
```

```python
sns.scatterplot(x= "Luas(m2)", y="Penduduk",data = sby2, color="blue")
```

```python
data_sby = sby2[['Luas(m2)','Penduduk']]
scaler = StandardScaler()
data_sby_scaled = scaler.fit_transform(data_sby)
data_sby_scaled.shape
```

```python
# Silhouette Analysis
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    
    # Initialise kmeans
    kmeans = KMeans(n_init=num_clusters, max_iter=50)
    kmeans.fit(data_sby_scaled)
    cluster_labels =  kmeans.labels_
    
    # Silhouette Score
    silhouette_avg = silhouette_score(data_sby_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
#iterasi menyesuaikan
scaler=StandardScaler()
wcc=[]
for i in range (1,10):
    km=KMeans(n_clusters=i)
    km.fit(scaler.fit_transform(sby2[['Luas(m2)','Penduduk']]))
    wcc.append(km.inertia_)
```

```python
plt.plot(range(1,10), wcc, marker='o')
```

```python
#init sesuai dengan pehitungan di atas
cluster=KMeans(n_clusters=3).fit_predict(scaler.fit_transform(sby2[['Luas(m2)','Penduduk']]))
sby2['KM4Cluster']=cluster
```

```python
plt.figure
sct = plt.scatter(x= "Luas(m2)", y="Penduduk",data = sby2, s = 100,
                  c = sby2.KM4Cluster, marker = "o", alpha = 0.5)
centers = kmeans.cluster_centers_
plt.scatter(centers[:,1], centers[:,0], c='blue', s=200, alpha=0.5);
plt.title("Hasil Klustering K-Means")
plt.xlabel("Scaled Longitude")
plt.ylabel("Scaled Latitude")
plt.show()
```

```python
plt.figure(figsize=[10,10])
sns.scatterplot(data = sby2, x= 'Luas(m2)',y= 'Penduduk' , hue = 'KM4Cluster' , palette="Set2", s = 100, alpha = 0.7)
```

```python
f, axes=plt.subplots(figsize=(10,10))
sby2.assign(cl=cluster).cx[:,:45].plot(column='cl', categorical=True, ax=axes, edgecolor='white', linewidth=0.1, alpha=0.7, legend=True)
axes.set_axis_off()
```

```python
sby3 = sby2.to_dict()
```

```python
panjang = len(sby3['KM4Cluster'])
```

```python
klasifikasi5 = list()
for i in range(0,panjang):
    if sby3['KM4Cluster'][i] == 0:
        klasifikasi5.append("cukup")
    elif sby3['KM4Cluster'][i] == 1:
        klasifikasi5.append("padat")
    else:
        klasifikasi5.append('renggang')
```

```python
klasifikasi5
```

```python
klas = gpd.GeoDataFrame(sby3)
klas['KM4Cluster']=klasifikasi5
```

```python
klas
```

```python
klas.plot(column='KM4Cluster', cmap='YlGn', figsize=(10,10), legend=True)
```

```python
klas['coords'] = klas['geometry'].apply(lambda x: x.representative_point().coords[:])
klas['coords'] = [coords[0] for coords in klas['coords']]

fig, ax = plt.subplots(figsize = (20,20))
klas.plot(ax=ax, column='KM4Cluster', categorical=True,   edgecolor='black',cmap='YlGn', legend=True)

for idx, row in klas.iterrows():
    plt.annotate(text=row['KECAMATAN'], xy=row['coords'], horizontalalignment='center')
```
